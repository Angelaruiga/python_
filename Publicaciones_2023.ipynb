```python
#@title Dar formato columna Authors
#de la selección de columnas que se realiza en el punto 6 Autoría selecionar_columnas = Autoria_Publicaciones2023[["Article Title", "Authors", 'DOI']]
df_publicaciones2023["Authors"] = df_publicaciones2023["Authors"].str.replace(", ", ",").str.replace("; ", ";")

# Función para quitar tildes y reemplazar 'ñ' por 'n'
def limpiar_texto(texto):
    # Reemplazar 'ñ' por 'n' y 'Ñ' por 'N'
    texto = texto.replace('ñ', 'n').replace('Ñ', 'N')
    texto = texto.replace('é', 'e').replace('É', 'E')
    texto = texto.replace('á', 'a').replace('Á', 'A')
    texto = texto.replace('í', 'i').replace('Í', 'I')
    texto = texto.replace('ó', 'o').replace('Ó', 'O')
    texto = texto.replace('ú', 'u').replace('Ú', 'U')


    # # Utilizar unidecode para quitar tildes y otros diacríticos
    # texto = unidecode.unidecode(texto)
    return texto

# Aplicar la función a la columna 'Authors'
df_publicaciones2023['Authors'] = df_publicaciones2023['Authors'].apply(limpiar_texto)
```

```python
# Importamos las bibliotecas necesarias.
import pandas as pd
import re

# Definimos la función `procesar_direcciones` que se encargará de procesar cada dirección.
def procesar_direcciones(address):
    # Comprobamos si la dirección es NA (No Disponible). En caso afirmativo, retornamos una cadena vacía.
    if pd.isna(address):
        return ""

    # Compilamos un patrón de expresión regular que usaremos para buscar en las cadenas de texto.
    # Este patrón busca grupos de texto que comienzan con "[" y terminan con "]", seguidos por cualquier texto hasta el próximo ";" o el final de la cadena.
    patron = re.compile(r'\[([^\]]+)\]\s*([^;]+)')

    # Inicializamos una lista vacía donde almacenaremos los fragmentos de texto procesados.
    resultado = []

    # Buscamos todas las coincidencias del patrón en la cadena `address`.
    # Cada coincidencia encontrada será una tupla que contiene los autores (dentro de corchetes) y su afiliación.
    matches = patron.findall(address)

    # Iteramos sobre cada tupla de autores y afiliaciones encontradas en la dirección.
    for autores, afiliacion in matches:
        # Dividimos la cadena de autores por "; " para manejar el caso de que haya múltiples autores listados juntos.
        autores_lista = autores.split('; ')

        # Iteramos sobre cada autor en la lista de autores.
        for autor in autores_lista:
            # Limpiamos el nombre del autor eliminando espacios en blanco extra al principio y al final de la cadena,
            # y reemplazamos ", " por "," para estandarizar el formato.
            autor_limpiado = autor.strip().replace(', ', ',')

            # Unimos el autor limpio con su afiliación usando "_" y agregamos esta cadena a la lista `resultado`.
            resultado.append(f"{autor_limpiado}_{afiliacion}")

    # Unimos todos los elementos de la lista `resultado` en una única cadena, separando cada elemento con "; ".
    # Esta cadena es lo que la función retorna.
    return '; '.join(resultado)

# Aplicamos la función `procesar_direcciones` a cada valor en la columna 'Addresses' de nuestro DataFrame `df`
# y almacenamos el resultado en una nueva columna llamada 'Processed_Addresses'.
Addresses_Publicaciones2023['Processed_Addresses'] = Addresses_Publicaciones2023['Addresses'].apply(procesar_direcciones)

# Imprimimos las columnas 'Addresses' y 'Processed_Addresses' del DataFrame para ver el resultado del procesamiento.
display(Addresses_Publicaciones2023[['Addresses', 'Processed_Addresses']])

```

```python
import pandas as pd

def check_intramural(df):
    # Crear la nueva columna 'Intramural' con valor por defecto 'No Intramural'
    df['Intramural'] = 'No Intramural'

    # Comprobar los artículos intramurales usando operaciones vectorizadas
    # Identificar filas donde el título del artículo es igual al de la fila anterior
    same_title = df['Article Title'].eq(df['Article Title'].shift())

    # Identificar filas donde el código de grupo es diferente al de la fila anterior
    different_code = df['Código Grupo'].ne(df['Código Grupo'].shift())

    # Aplicar ambas condiciones para marcar como 'Intramural'
    intramural_mask = same_title & different_code
    df.loc[intramural_mask, 'Intramural'] = 'Intramural'
    df.loc[intramural_mask.shift(-1, fill_value=False), 'Intramural'] = 'Intramural'

    return df

# Suponiendo que df_intramural es tu DataFrame ya definido
df_intramural = check_intramural(df_intramural)

# Mostrar algunas filas para verificar los resultados
print(df_intramural[['Article Title', 'Código Grupo', 'Intramural']].head())
```
