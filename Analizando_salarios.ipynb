```python
# Importa librerías esenciales
import pandas as pd
import matplotlib.pyplot as plt

# Carga tu dataset
# Por simplicidad, usaremos un dataset de ejemplo de pandas
data = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'María', 'Juan', 'Carlos'],
    'Edad': [23, 35, 29, 42, 34],
    'Salario': [50000, 60000, 58000, 62000, 70000]
})

# Realiza análisis básicos
edad_media = data['Edad'].mean()
salario_medio = data['Salario'].mean()

# Visualiza tus resultados
plt.figure(figsize=(10, 5))
plt.bar(data['Nombre'], data['Salario'], color='skyblue')
plt.xlabel('Nombre')
plt.ylabel('Salario')
plt.title('Salario por Persona')
plt.axhline(salario_medio, color='red', linestyle='--', label=f'Salario Medio: ${salario_medio:.2f}')
plt.legend()
plt.show()

print(f"Edad media: {edad_media:.2f} años")
print(f"Salario medio: ${salario_medio:.2f}")
```
