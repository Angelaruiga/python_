## ¿Influye el Modo de Envios en la tasa de Devolución?

En el excel de tenemos conocimiento sobre devoluciones. Puedes "unirlo" (merge) a nuestro dataset para ver si la elección del metodo de envio puede repercutir en las devoluciones.

Preparación del dato:

```python
#creamos un nuevo DF para traernos la columna que queremos del excel
df_returns = pd.read_excel("PontiaSuperStore.xls", sheet_name="Returns")
```

```python
num_nan = df_returns['Returned'].isna().sum()
print(f"La columna 'Returned' tiene {num_nan} valores NaN.")
```

```python
df_returns['Returned'].value_counts()
```

```python
#union de los dos df...pondriamos on='Primary Key' en el caso que fueramos a unir por clave primaria

merged_df = df.merge(df_returns, how='left')
```

```python
#comprobamos
merged_df.dtypes
```

```python
merged_df.head() #normal que tenga NaN entendemos por esto que no se ha devuelto
```

```python
num_filas, num_columnas = merged_df.shape
print(f"El DataFrame tiene {num_filas} filas y {num_columnas} columnas.")
```

```python
num_nan = merged_df['Returned'].isna().sum()
print(f"La columna 'Returned' tiene {num_nan} valores NaN.")
```

```python
nan= round((num_nan/num_filas)*100,2)
print(f" El {nan} % de los valores en 'Returned'son NaN.")
```

```python
#cambiar los NaN por No
merged_df['Returned'] = merged_df['Returned'].fillna('No')
#haciendo el paso de abajo este codigo no sería necesario,
```

```python
#crear una nueva columna para "Devuelvo" y "No devuelto"
merged_df['flag_returned'] = np.where(merged_df['Returned'] == 'Yes', 'Devuelto', 'No Devuelto').astype(object)
```

```python
# Agrupar los datos por 'Ship Mode' y calcular el número total de pedidos y devoluciones para cada grupo con la funcion lambda
df_return = merged_df.groupby('Ship Mode').apply(lambda x: pd.Series({
    'Total Orders': len(x),
    'Returns': (x['flag_returned'] == 'Devuelto').sum()
})).reset_index()

# Calcular la tasa de devolución para cada 'Ship Mode'
df_return['Return Rate'] = round((df_return['Returns'] / df_return['Total Orders']) * 100, 2)
df_return
```

```python
# Crear el gráfico de barras con Plotly
fig_strip_return = px.bar(df_return, x='Ship Mode', y='Return Rate')
fig_strip_return.update_layout(
    xaxis_title="Modo de envío",
    yaxis_title="Tasa de devolución",
    font = {
        'family':'opensans, monospace',
        'size':12,
        'color':'black'
        },
    title={
        'text': '<b>Tasa de devolución según envío</b>',
        'y':0.99,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'})
fig_strip_return.update_xaxes(showgrid=False)  #False desaparece la línea
fig_strip_return.update_yaxes(showgrid=True)
fig_strip_return.update_yaxes(zeroline=True, zerolinewidth=2, zerolinecolor='grey')
fig_strip_return
```

Respuesta: Se observa que el modo de envío  podría influir en la tasa de devolución. Sobre todo en los envíos “First Class” y “Same Day” ya que muestran tasas de devolución más altas en comparación con “Second Class” y “Standard Class”.

Una posible hipótesis podría ser que la expectativa del cliente sobre el producto es más alta en los envíos de entrega rápida, por ejemplo. Pero esto se necesitaría de otros análisis para argumentar y comprobar esta hipotesis.
