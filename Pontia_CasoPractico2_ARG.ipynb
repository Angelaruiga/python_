# **Tipos de Envios**

## ¿Qué porcentaje de las ventas se hacen según cada tipo de Modalidad de envio por Barco?

#### Preparación del Dato.

Realiza las agregaciones necesarias para poder graficar.

Respuesta:

```python
df_pie = df.groupby('Ship Mode', as_index=False).count()[['Ship Mode', 'Sales']]
df_pie
```

Elegimos un gráfico de...


Queremos que quede para enseñarselo al mismisimo CEO!
- Customiza el Titulo.
- El nombre de las etiquetas.
- Queremos que de forma interactiva aparezca el nº Ventas.

Respuesta:

```python
# Grafico de Tarta.

fig_pie_shipmode=px.pie(df_pie, names='Ship Mode', values='Sales', hole=0.3, color_discrete_sequence=['darkred', '#27AE60', '#0080FF', '#FF8C00'] ,opacity = 0.75)
fig_pie_shipmode.update_traces(marker = dict(line = dict(color = 'white', width = 2)))
fig_pie_shipmode.update_traces(textinfo='label+percent') #incluir las etiquetas dentro del gráfico
fig_pie_shipmode.update_layout(showlegend=False, #indicar que no muestre la leyenda
    title={
        'text': '<b>% de ventas según la modalidad de envio</b>', #poner en negrita el título
        'y':0.99,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font=dict(size=16, color='black', family='opensans, monospace'))
```

## ¿Cómo han evolucionado con el tiempo las modalidades de envios?

Estudiando la evolución temporal queremos ver si es algo que se ha mantenido desde el origen, si ha aumentado o decrecido con el paso del tiempo...

Preparación del Dato:

```python
# Preparacion del dato.
df_line = df.groupby(['Ship Mode','Year'], as_index=False).count()[['Ship Mode','Sales','Year']]
df_line
```

Respuesta:

Eligiremos una gráfica de...

```python
# Plotly Express para Gráficar una linea.
df_line['Year'] = df_line['Year'].astype(str)
fig_line_shipmode = px.line(df_line, x='Year', y='Sales', color='Ship Mode',
color_discrete_map=colores_envios, line_shape='spline', labels={'Year': 'Año', 'Sales': 'Ventas', 'Ship Mode':'Modo de envío'})
fig_line_shipmode.update_layout(
    title={
        'text': '<b>Evolución modalidad de envío</b>',
        'y':0.99,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font=dict(size=16, color='black', family='opensans, monospace'))
fig_line_shipmode
```

Elementos de interes a añadir en la gráfica:

- Añade Grid en la gráfica (fig_line_discount.update_xaxes)
- Añade el importe total de las ventas en hover_data (quizá tengas que modificar el groupby...)

```python
#Group by para conocer el total de ventas
df_line_2 = df.groupby(['Ship Mode','Year'], as_index=False)['Sales'].sum()
```

```python
# Plotly Express para Gráficar una linea.
df_line_2['Year'] = df_line_2['Year'].astype(str)
fig_line_shipmode_2 = px.line(df_line_2, x='Year', y='Sales', color='Ship Mode',color_discrete_map=colores_envios, line_shape='spline', labels={'Year': 'Año', 'Sales': 'Ventas', 'Ship Mode':'Modo de envío'},hover_data=['Sales'])
fig_line_shipmode_2.update_layout(
    title={
        'text': '<b>Evolución modalidad de envío</b>',
        'y':0.99,
        'x':0.48,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font=dict(size=16, color='black', family='opensans, monospace'))
fig_line_shipmode_2.update_xaxes(showgrid=False)  #False desaparece la línea
fig_line_shipmode_2.update_yaxes(showgrid=True)
fig_line_shipmode_2
```
