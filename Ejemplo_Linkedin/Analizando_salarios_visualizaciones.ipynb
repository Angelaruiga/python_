```python
# Importa librerías esenciales
import pandas as pd  # Importa la librería pandas para el manejo de datos
import matplotlib.pyplot as plt  # Importa matplotlib para la visualización de datos
import seaborn as sns  # Importa seaborn para visualizaciones estadísticas avanzadas
import plotly.express as px  # Importa plotly.express para visualizaciones interactivas
```

```python

# Carga tu dataset
data = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'María', 'Juan', 'Carlos'],  # Nombres de las personas
    'Edad': [23, 35, 29, 42, 34],  # Edades correspondientes
    'Salario': [50000, 60000, 58000, 62000, 70000]  # Salarios correspondientes
})
```

```python

# Realiza análisis básicos
edad_media = data['Edad'].mean()  # Calcula la edad media
salario_medio = data['Salario'].mean()  # Calcula el salario medio
```

```python

#Gráfico 1
# Visualiza tus resultados con matplotlib
plt.figure(figsize=(10, 5))  # Configura el tamaño de la figura
plt.bar(data['Nombre'], data['Salario'], color='skyblue')  # Crea una gráfica de barras
plt.xlabel('Nombre')  # Etiqueta del eje X
plt.ylabel('Salario')  # Etiqueta del eje Y
plt.title('Salario por Persona')  # Título del gráfico
plt.axhline(salario_medio, color='red', linestyle='--', label=f'Salario Medio: €{salario_medio:.0f}')  # Añade una línea horizontal indicando el salario medio
plt.legend()  # Muestra la leyenda
plt.show()  # Muestra la gráfica

print(f"Edad media: {edad_media:.0f} años")  # Imprime la edad media, redondeada al entero más cercano
print(f"Salario medio: €{salario_medio:.0f}")  # Imprime el salario medio, redondeado al entero más cercano

#Gráfico 2

# Visualiza tus resultados con seaborn
plt.figure(figsize=(10, 5))  # Configura el tamaño de la figura
sns.barplot(x='Nombre', y='Salario', data=data, hue='Nombre', palette='viridis', dodge=False)  # Crea una gráfica de barras
plt.xlabel('Nombre')  # Etiqueta del eje X
plt.ylabel('Salario')  # Etiqueta del eje Y
plt.title('Salario por Persona')  # Título del gráfico
plt.axhline(salario_medio, color='red', linestyle='--', label=f'Salario Medio: €{salario_medio:.0f}')  # Añade una línea horizontal indicando el salario medio
plt.legend().set_visible(False)  # Oculta la leyenda para mayor claridad
plt.show()  # Muestra la gráfica

print(f"Edad media: {edad_media:.0f} años")  # Imprime la edad media, redondeada al entero más cercano
print(f"Salario medio: €{salario_medio:.0f}")  # Imprime el salario medio, redondeado al entero más cercano

#Gráfico 3

# Visualiza tus resultados con Plotly
fig = px.bar(data, x='Nombre', y='Salario', title='Salario por Persona', labels={'Salario':'Salario'}, text='Salario')  # Crea una gráfica de barras interactiva
fig.add_shape(type='line',
              x0=-0.5, x1=4.5, y0=salario_medio, y1=salario_medio,
              line=dict(color='Red', dash='dash'))  # Añade una línea horizontal indicando el salario medio

# Actualiza la configuración del gráfico
fig.update_layout(
    shapes=[dict(
        type="line",  # Tipo de forma: línea
        xref="paper",  # Referencia en el eje X: papel (el rango de toda la figura)
        x0=0, x1=1,  # Coordenadas en el eje X (de 0 a 1, es decir, a lo largo de toda la gráfica)
        yref="y",  # Referencia en el eje Y: datos (el rango de datos del eje Y)
        y0=salario_medio, y1=salario_medio,  # Coordenadas en el eje Y: posición de la línea horizontal
        line=dict(color="Red", dash="dash"),  # Estilo de la línea: color rojo y guiones
    )],
    annotations=[dict(
        x=0.5,  # Posición en el eje X para la anotación (0.5 es el medio de la gráfica)
        y=salario_medio,  # Posición en el eje Y para la anotación: el valor del salario medio
        xref="paper",  # Referencia en el eje X: papel (el rango de toda la figura)
        yref="y",  # Referencia en el eje Y: datos (el rango de datos del eje Y)
        text=f'Salario Medio: €{salario_medio:.2f}',  # Texto de la anotación
        showarrow=False,  # No mostrar una flecha apuntando al punto de la anotación
        font=dict(color="Red"),  # Color de la fuente de la anotación: rojo
        align='left',  # Alineación del texto de la anotación: izquierda
        bgcolor="white",  # Fondo blanco para la anotación
        bordercolor="red",  # Color del borde de la anotación: rojo
        borderwidth=1  # Ancho del borde de la anotación
    )]
)

fig.show()  # Muestra la gráfica interactiva

print(f"Edad media: {edad_media:.0f} años")  # Imprime la edad media, redondeada al entero más cercano
print(f"Salario medio: €{salario_medio:.0f}")  # Imprime el salario medio, redondeado al entero más cercano
```
