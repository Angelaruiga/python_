```python
# Mostrar los nombres de todas las columnas
print("\nNombres de las columnas:")
print(df.columns)
```

```python


# Agrupar por el título del artículo y el nombre de la revista, concatenando los códigos de grupo únicos
df_grouped = df.groupby(['Article Title', 'ource Title']).agg({
    'Authors': lambda x: ', '.join(sorted(set(x))),
    'Publication Year': 'first',
    'Volume': 'first',
    'Issue': 'first',
    'Start Page': 'first',
    'End Page': 'first',
    'Special Issue': 'first',
    'Part Number': 'first',
    'DOI': 'first',
    'Código Grupo': lambda x: ', '.join(sorted(set(x))),
    'D1': 'first',  # Si D1 tiene un solo valor por grupo, usar 'first' para tomar ese valor
    'Quartiles': lambda x: ', '.join(sorted(set(x))),
    'Intramural_agrupado.Intramural': lambda x: ', '.join(sorted(set(map(str, x)))),
    'Artículos': 'count'  # Contar el número de artículos en el grupo
}).reset_index()

# Mostrar una vista previa de los datos procesados
print(df_grouped.head())

# Guardar el DataFrame en un archivo Excel
output_file = 'publicaciones_2023_agrupadas.xlsx'
df_grouped.to_excel(output_file, index=False)

# Descargar el archivo Excel (en caso de estar en Google Colab)
from google.colab import files
files.download(output_file)
```

```python
import pandas as pd
from openpyxl import Workbook
from google.colab import files

# Cargar el archivo Excel
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)
```

```python

# Crear una nueva columna con el formato deseado
df['Formatted'] = df.apply(lambda row: f"{row['Article Title']}. {row['Authors']}. {row['Publication Year']}. \
{row['ource Title']}. {row['Volume']}. {row['Issue']}. {row['Start Page']}-{row['End Page']}. \
{row['DOI']}", axis=1)

# Crear un nuevo DataFrame solo con la columna formateada
df_formatted = df[['Formatted']]

# Guardar el nuevo DataFrame en un archivo Excel
output_file = 'publicaciones_formateadas.xlsx'
df_formatted.to_excel(output_file, index=False)

# Descargar el archivo Excel
files.download(output_file)
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
from google.colab import files

# Subir el archivo Excel
uploaded = files.upload()


# Leer el archivo Excel especificando la hoja y la fila de los nombres de las columnas
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name, sheet_name='DESCARGADO 100524') # Ajusta el nombre de la hoja, header=1)  # Ajusta el índice según corresponda
```

```python
# Mostrar los nombres de todas las columnas
print("\nNombres de las columnas:")
print(df.columns)
```

```python
# Seleccionar las columnas necesarias
df_selected = df[['Código', 'Título', 'Nombre IP', 'Nombre CO IP', 'Tipo Conv', 'Tipo Financiación', 'Naturaleza']]

# Filtrar por 'Tipo Conv' que contenga 'PI', 'RH-PI' o 'AC-PI'
df_filtered = df_selected[df_selected['Tipo Conv'].str.contains('PI|RH-PI|AC-PI', na=False, case=False)]

# Filtrar por 'Tipo Financiación' que contenga 'SUBVENCION'
df_filtered = df_filtered[df_filtered['Tipo Financiación'].str.contains('SUBVENCION', na=False, case=False)]

# Filtrar por 'Naturaleza' que contenga 'SUBVENCION'
df_filtered = df_filtered[df_filtered['Naturaleza'].str.contains('PUBLICO', na=False, case=False)]


# Ordenar por 'Nombre IP'
df_filtered = df_filtered.sort_values(by='Nombre IP')

# Convertir 'Título' a tipo oración (capitalizar la primera letra)
df_filtered['Título'] = df_filtered['Título'].str.capitalize()

# Reemplazar NaN en 'Nombre IP' y 'Nombre CO IP' con una cadena vacía
df_filtered['Nombre IP'] = df_filtered['Nombre IP'].fillna('')
df_filtered['Nombre CO IP'] = df_filtered['Nombre CO IP'].fillna('')

# Crear una nueva columna con el formato deseado
df_filtered['Formato'] = df_filtered['Título'] + '. ' + df_filtered['Código'] + '. ' + df_filtered['Nombre IP'] + ', ' + df_filtered['Nombre CO IP']

# Seleccionar la nueva columna para exportar
df_formatted = df_filtered[['Formato']]

# Guardar el resultado en un nuevo archivo Excel
output_file = 'proyectos_formato.xlsx'
df_formatted.to_excel(output_file, index=False)

# Descargar el archivo Excel
files.download(output_file)
```
