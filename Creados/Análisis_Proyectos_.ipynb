```python
#@title Verificar  FECHA DE CONCESI√ìN

# 1. Fecha de inicio y fecha fin est√°n completas pero falta fecha de concesi√≥n
falta_fecha_concesion = df[(df['Fecha_Inicio'].notna()) & (df['Fecha_Fin'].notna()) & (df['Fecha_Concesion'].isna())]

# 2. Fecha de inicio est√° completa y fecha fin falta, pero la fecha de concesi√≥n est√° completa
falta_fecha_fin = df[(df['Fecha_Inicio'].notna()) & (df['Fecha_Fin'].isna()) & (df['Fecha_Concesion'].notna())]


# Mostrar los resultados
print("Casos donde falta la fecha de concesi√≥n:\n", falta_fecha_concesion[['Codigo', 'Ano_Convocatoria', 'Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion']])
print("\nCasos donde falta la fecha fin:\n", falta_fecha_fin[['Codigo', 'Ano_Convocatoria', 'Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion']])
```

```python
#@title Verificar  ESTADO


# Definir campos relevantes para el chequeo
campos_relevantes = ['Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion', 'Total_Concedido']

# Casos donde el estado es DENEGADO y deber√≠an estar vac√≠os, pero no lo est√°n
denegado_incorrecto = df[(df['Estado'] == 'DENEGADO') &
                         (df[campos_relevantes].notna().any(axis=1)) &
                         (df['Total_Concedido'] != 0)]

# Casos donde el estado es SOLICITADO y deber√≠an estar vac√≠os, pero no lo est√°n
solicitado_incorrecto = df[(df['Estado'] == 'SOLICITADO') &
                           (df[campos_relevantes].notna().any(axis=1)) &
                           (df['Total_Concedido'] != 0)]

# Si ESTADO es ACTIVO, todos los campos deben estar completos
activo_incorrecto = df[(df['Estado'] == 'ACTIVO') &
                       (df[['Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion', 'Total_Solicitado', 'Total_Concedido']].isna().any(axis=1))]

# Mostrar los resultados
print("Casos donde el estado es ACTIVO pero falta alg√∫n campo:\n", activo_incorrecto[['Codigo', 'Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion', 'Total_Solicitado', 'Total_Concedido']])
print("\nCasos donde el estado es DENEGADO pero algunos campos requeridos est√°n completos:\n", denegado_incorrecto[['Codigo', 'Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion', 'Total_Concedido']])
print("\nCasos donde el estado es SOLICITADO pero algunos campos requeridos est√°n completos:\n", solicitado_incorrecto[['Codigo', 'Fecha_Inicio', 'Fecha_Fin', 'Fecha_Concesion', 'Total_Concedido']])

```

# An√°lisis Entidad Financieadora y Concesiones

```python

# Agrupar por las columnas especificadas y calcular el n√∫mero de concedidos y el total concedido
grupo = df_publico.groupby(['Nombre_Entidad_Financiadora_Abreviada', 'a√±o_Convocatoria','a√±o_concesi√≥n', 'Tipo_Conv', 'Nombre_Empresa']).agg(
    Numero_Concedidos=('concedidos', 'sum'),
    Total_Concedido=('Total_Concedido', 'sum')
).reset_index()
```

```python
# Crear una tabla pivot con un nombre m√°s espec√≠fico
pivot_concesiones_entidad = grupo.pivot_table(
    index=['Nombre_Entidad_Financiadora_Abreviada', 'a√±o_Convocatoria', 'a√±o_concesi√≥n'],
    columns='Tipo_Conv',
    values=['Numero_Concedidos', 'Total_Concedido'],
    aggfunc='sum',
    fill_value=0
)

# Mostrar la tabla pivot
print("Tabla Pivot:")
display(pivot_concesiones_entidad)

# Guardar el DataFrame en un archivo Excel con un nombre m√°s espec√≠fico
output_file_path = "/content/drive/My Drive/Proyectos a corregir/Proyectos_Concesiones_Entidades.xlsx"
pivot_concesiones_entidad.to_excel(output_file_path)

print("Archivo guardado con √©xito en:", output_file_path)
```
